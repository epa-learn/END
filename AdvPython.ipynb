```
fibonacci_numbers = [0, 1]
while fibonacci_numbers[-1] + fibonacci_numbers[-2] < 10000:
    fibonacci_numbers.append(fibonacci_numbers[-1] + fibonacci_numbers[-2])
fibonacci_numbers    
check_fib = lambda x : 'True' if x in fibonacci_numbers else 'False'
check_fib(21)    
```

```
def add_2_iterables(l1: list, l2: list):
  """
  Using list comprehension 
  add 2 iterables a and b such that a is even and b is odd
  """
  return sum([a + b for a, b in zip(l1, l2) if a % 2 == 0 and b % 2 != 0])

add_2_iterables([2,4,6,8,9],[2,3,4,5,6])
```

```
vowels = ['a', 'e', 'i', 'o', 'u']
stripVowel= lambda s : [x for x in s if x not in vowels]
stripVowel('tsai')
    
```

```
relu_func = lambda l :[x if x > 0 else 0 for x in l]
relu_func([-0.2, 5])
```

```
#siggmoid function
sigmoid_func = lambda l : [(1 / (1 + math.exp(-x))) for x in l]
sigmoid_func([2,-9.7])
```

```
# Reduce function to find the biggest character in a string
find_biggest_char = lambda str : reduce(lambda x, y: x if ord(x) > ord(y) else y, str)    
find_biggest_char("Sachin")
```
